```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
dataset
```

```python
x = pd.DataFrame(data = dataset.data,columns=dataset.feature_names)
x
```

```python
y = pd.DataFrame(data = dataset.target)
y
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state=56)
```

```python
# pip install mlxtend
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
EFS = EFS(lr,max_features = 4,scoring = 'accuracy',cv=10)
# max_features represents the maximum nimber of features that we want in our model.
# scroing parameter is used to specify the scoring method that we want to use to evaluate the model.
# cv parameter is used to specify the number of folds that we want to use in cross validation. This time we are using 10 cvs,  that means we'll be finding the accuracy of our model 10 times and we'll be taking the average of those 10 accuracies.
```

```python
model = EFS.fit(x_train,y_train)
```

#### we can see above that the total number of subsets formed are 15, which comes from (2^n -1) , here n = 4

```python
model.best_feature_names_
```

```python
model.best_score_
```

```python
subsets = model.subsets_
subsets
```

```python
subsets = pd.DataFrame(subsets).T
subsets
```

```python
import matplotlib.pyplot as plt
plt.plot([str(k) for k in subsets['feature_names']],subsets['avg_score'])
plt.xticks(rotation = 90)
```
