```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('BostonHousing.csv')
dataset
```

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
print(np.mean(cross_val_score(model,x_train,y_train,cv = 5, scoring = 'r2')))
print(np.mean(cross_val_score(model,x_test,y_test,cv = 5, scoring = 'r2')))
```

#### We can clearly see that there is a large gap in explainablity of variance in training and testing data. so, we can say that there is an overfitting in model.

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector
EFS = ExhaustiveFeatureSelector(model,max_features = 13, scoring = 'r2',cv = 10, n_jobs = -1)
sel = EFS.fit(x_train,y_train)
```

```python
sel.best_feature_names_
```

#### It is giving numbers as, the column names were removed as we performed the standardization and dataframe was converted into numpy array.

```python
sel.best_score_
```

```python
x_sel = sel.subsets_
x_sel = pd.DataFrame(x_sel).T
x_sel.sort_values('avg_score',ascending=False)
# x_sel
```

```python
def AdjR2(r2,n,k):
    return 1 - ((1-r2)*(n-1)/(n-k-1))
```

```python
x_sel['observations'] = 404
x_sel['no. of features'] = x_sel['feature_idx'].apply(lambda x: len(x))
x_sel['Adjusted R2'] = AdjR2(x_sel['avg_score'],x_sel.shape[0],x_sel['no. of features'])
x_sel.sort_values('Adjusted R2',ascending = False)
# x_sel
```

#### Here, the first row has the maximum Adusted R2 score which means that the set of features (0, 1, 4, 5, 7, 8, 9, 10, 11, 12) will be able to explain the variability in dependent variable best.
