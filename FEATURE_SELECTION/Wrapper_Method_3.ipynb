```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('BostonHousing.csv')
dataset
```

```python
x = dataset.iloc[:,:-1]
x
```

```python
y = dataset.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_train
```

```python
x_test = sc.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.model_selection import cross_val_score
print("Training:", np.mean(cross_val_score(model,x_train,y_train,cv=10,scoring='r2')))
print("Testing:", np.mean(cross_val_score(model,x_test,y_test,cv=10,scoring='r2')))
```

#### We can clearly see that there is a large gap in explainablity of variance in training and testing data. so, we can say that there is an overfitting in model.

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
sfs = SFS(model,k_features='best',forward = False,floating = False, scoring = 'r2',cv = 10)
```

```python
sel = sfs.fit(x_train,y_train)
```

```python
sel.k_feature_names_
```

```python
x_sel = sel.subsets_
x_sel = pd.DataFrame(x_sel).T
x_sel
```

```python
def Adjusted_R2(r2,n,k):
    return 1- ((1-r2)*(n-1)/(n-k-1))
```

```python
x_sel['total rows'] = x_train.shape[0]
x_sel['no. of features'] = x_sel['feature_idx'].apply(lambda x:len(x))
x_sel['Adjusted R2'] = Adjusted_R2(x_sel['avg_score'],x_sel['total rows'],x_sel['no. of features'])
x_sel.sort_values('Adjusted R2',ascending=False)
```

```python
x_train = sel.transform(x_train)
```

```python
x_train = pd.DataFrame(x_train)
x_train
```

```python
x_test = sel.transform(x_test)
x_test = pd.DataFrame(x_test)
x_test
```

```python
model.fit(x_train,y_train)
```

```python
y_pred_2 = model.predict(x_test)
y_pred_2
```

```python
print("Training:", np.mean(cross_val_score(model,x_train,y_train,cv=10,scoring='r2')))
print("Testing:", np.mean(cross_val_score(model,x_test,y_test,cv=10,scoring='r2')))
```
