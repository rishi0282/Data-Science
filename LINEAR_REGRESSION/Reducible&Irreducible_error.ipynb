```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# we have randomly generated some values for x, which will serve as the independent variable
x = np.random.rand(50) * 10
x
```

```python
# Let's say we already now the exact actual parameter for the population dataset.
# Also, we have some generated irreducible error by np.random.rand(50) * 4
irred = np.random.rand(50) * 4  # multipling it to make it a bit larger number
y = 2 * x - 5  + irred
y
```

```python
x = x.reshape(-1,1)
x
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
y_pred = lr.predict(x)
y_pred
```

```python
plt.scatter(x,y,label="data point")

# plot actual line 
x_line = np.linspace(0,10,100)
y_actual = 2 * x_line - 5
# print(x_line)
plt.plot(x_line,y_actual,color = 'green',label = 'population line (m=2,b=-5)')
# plot regression line
y_estimated = lr.coef_[0] * x_line + lr.intercept_
plt.plot(x_line,y_estimated,color='purple',label = f"Estimated line, m:{lr.coef_[0]:.2f},b:{lr.intercept_:.2f}")
plt.legend()
```
