#### Batch Gradient Descent 

```python
class GDregressor:
    def __init__(self,learning_rate,epochs):
        self.lr = learning_rate
        self.epochs = epochs
        self.coeff_ = None
        self.intercept_ = None

    def fit(self,x_train,y_train):
        # Whenever we start with the gradient descent we start with random values of each parameter and then update them at each iteration.
        # So, In this case we are intializing them as 1.
        """ This is a function which trains on Linear Regression for the given dataset and use Gradient Descent as its Loss Optimization technique.\n Parameters: \n x_train = input variables of the training dataset,\n y_train = Output variables of the train dataset """
        self.coeff_ = np.ones(x_train.shape[1])
        self.intercept_ = 0      # B0 
        # print(f"shape of coeff_ : {self.coeff_.shape}")
        for i in range (self.epochs):
            # We calculated y_pred using the dot product method (m x n) . (n x 1)
            y_pred = np.dot(x_train,self.coeff_) + self.intercept_
            self.intercept_ = self.intercept_ - (self.lr * np.mean(y_train - y_pred) * (-2))
            self.coeff_ = self.coeff_ - (self.lr * -2 * np.dot(y_train-y_pred,x_train)/x_train.shape[0])

        print(self.intercept_)
        # print("shape: ",self.coeff_.shape)
        print(self.coeff_)

    def predict(self,x_test):
        y_pred = self.intercept_ + np.dot(x_test,self.coeff_)
        # print(y_pred)
        return y_pred
```

#### Mini Batch Gradient Descent 

```python
class MBGDregressor:
    def __init__(self,learning_rate,epochs,batch_size):
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.coeff_ = None
        self.intercept_ = None

    def fit(self,x_train,y_train):
        # Whenever we start with the gradient descent we start with random values of each parameter and then update them at each iteration.
        # So, In this case we are intializing them as 1.
        """ This is a function which trains on Linear Regression for the given dataset and use Mini Batch Gradient Descent as its Loss Optimization technique.\n Parameters: \n x_train = input variables of the training dataset,\n y_train = Output variables of the train dataset """


        self.coeff_ = np.ones(x_train.shape[1])
        self.intercept_ = 0      # B0 
        
        for i in range (self.epochs):
            for j in range ( int(x_train.shape[0]/self.batch_size) ):

                # We are choosing the batch_size number of random indexes for which we will conduct our one update
                #i.e. let's say we are choosing 10 random values which will be used to update the intercept and variable 
                index = random.sample(range(0,x_train.shape[0]),self.batch_size)

                y_pred = np.dot(x_train[index],self.coeff_) + self.intercept_
                intercept_derivative = np.mean(y_train[index] - y_pred) * (-2)
                self.intercept_ = self.intercept_ - self.lr * intercept_derivative

                coeff_derivative = (np.dot(y_train[index] - y_pred,x_train[index]) * (-2)) / (self.batch_size)
                self.coeff_ = self.coeff_ - self.lr * coeff_derivative

        print(self.intercept_)
        print(self.coeff_)

    def predict(self,x_test):
        y_pred = np.dot(x_test,self.coeff_) + self.intercept_
        return y_pred
```

#### For MBGDRegressor

```python
learning_rate = 0.05
epochs = 100
batch_size = 50
mbgr = MBGDregressor(learning_rate,epochs,batch_size)
mbgr.fit(x_train,y_train)
```

```python
y_pred = mbgr.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2)
n = y_test.shape[0]
k = x_train.shape[1]
print(f"n: {n}, k: {k}")
r2_adj = 1 - ((1-r2)*(n-1)/(n-1-k))
r2_adj
```
